```python
from PIL import Image

import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
device = torch.device('cuda:0') if torch.cuda.is_available() else torch.device('cpu')
print(device)
```

```python
Image.MAX_IMAGE_PIXELS = None
image1 = Image.open('../Database/Clipping_rescale/PR.png')
image2 = Image.open('../Database/Clipping_rescale/PG.png')
image3 = Image.open('../Database/Clipping_rescale/PB.png')
image4 = Image.open('../Database/Clipping_rescale/PN.png')
```

```python
# Open the images
images = [image1, image2, image3]

# Convert images to NumPy arrays
arrays = [np.array(image) for image in images]

# Stack the arrays along the last axis (axis=-1) to create a single array
merged_array = np.stack(images, axis=-1)

# merged_array = merged_array.astype(np.float32) / 255.0

# Create an image from the merged NumPy array
merged_image = Image.fromarray(merged_array, 'RGB')
```

```python
# Find the maximum pixel value
max_pixel_value = np.amax(merged_image)

print("Maximum pixel value:", max_pixel_value)
```
